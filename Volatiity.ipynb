### **DEFINITION OF VOLATILITY RISK**
Volatility risk is the risk of a change of price of a portfolio as a result of changes in the volatility of a risk factor. It usually applies to portfolios of derivatives instruments, where the volatility of its underlying is a major influencer of prices. This is the risk to the value of an investment, due to unpredictable changes in the volatility of the underlying asset. Volatility risk is specifically related to the breadth of the trading range between the high and low price levels at which a stock or commodity has traded.

Volatility risks are those that are associated with the movement of an investment's price from one point in time to another. These risks can include the potential that the price of the security will change suddenly and in an unpredictable manner, as well as the possibility that the price of the security could move in a direction that is opposite to what was predicted for its movement. The relationship that exists between two or more stocks might bring certain concerns, often known as correlation hazards. These dangers can be potentially catastrophic for investors. These risks can include the possibility that the values of the securities will move in the same direction, or they can include the probability that the values of the securities will move in the opposite direction. Both of these outcomes are possible.

Volatility and correlation are two indicators that are used in the world of investments for the aim of gauging risk. Both of these indicators are used in conjunction with one another. For instance, a potential investor might choose to put their money into a stock that has both a low volatility and a significant connection with the entire stock market. This would be advantageous for the potential investor in several ways.

It is worthy to note that, if the price of a security fluctuates in a short time span, it is termed to have high volatility. However, if the price of a security fluctuates slowly over a longer time span, it is termed to have low volatility.

For the purpose of this work, we'll focus on Volatility seperately from correlation.

### **WHAT IS VOLATILITY?**

Volatility is a measure of how much the price of an asset fluctuates over time. It concerns how quickly the price changes, how often it changes, and how much. Volatility is thought of by many as the kissing cousin of risk. For a short-term trader, market volatility offers the possibility of bigger gains and the risk of bigger losses. Speculative traders embrace volatility as the main engine of their profits.

In the bigger picture, volatility involves asset price changes and not asset quantity. And its important to understand that price and value are not identical. A stocks price might fluctuate wildly, but the companys value can remain unchanged. What has changed is the price market participants are willing to pay to own that intrinsic value, and prices swing significantly depending on market conditions.

For long-term investors, volatility is much less of a concern. Short-term price fluctuations tend to even out over time, which is why investors with long time horizons typically invest more of their portfolios in stocks. As the time horizon shortens and the investor gets closer to retirement, the allocation to equity is reduced while investment in bonds is increased. Bonds are much less volatile than stocks, although they carry varying levels of risk.

### **TYPES OF VOLATILITY**

The generally believed types of volatility are:

1. **Historical Volatility:** This measures the fluctuations in the securitys prices in the past. It is used to predict the future movements of prices based on previous trends. However, it does not provide insights regarding the future trend or direction of the securitys price. Historical volatility is the standard deviation of the annualised daily returns of the underlying. Generally, it is calculated on a rolling basis to keep a recency bias.

2. **Implied Volatility:** This is a future projection of probable movements in values of securities. Implied volatility helps to determine the future of a particular stocks value, without taking into account its historical data. This is an essential metric that helps determine the prices of options contracts. Implied volatility is expressed in percentages, however, it does not clarify the direction in which prices will move.

    ![Implied Volatibity and Historical Volatility](implied-vol.png)



A high implied volatility suggests that the price of a specific security will change dramatically -- it could either appreciate or depreciate in value.  On the other hand, a low implied volatility suggests that the price of a specific security  will not change dramatically, apart from minor deviations.

Therefore, a bearish market will show a high implied volatility rate in comparison to a bullish market, where implied volatility will be low. This is because in a bearish market, the prices are predicted to decline over time, whereas in a bullish market, the prices are expected to increase over time.

### **EXAMPLE OF VOLATILITY OF TWO ASSET CLASSES WITH REAL-WORLD DATA**

The two asset used in analysis are:
1. ISFD.L: iShares Core FTSE 100 UCITS ETF.
2. IGLN.L: iShares Physical Gold ETC.

Both are traded in the US.

#### **Importing Data**

```python


import yfinance as yf
import pandas as pd
import numpy as np

import datetime as dt
from datetime import date
```

### **Calculating Daily Returns**

```python
# Describing the dataset
tickers = ["ISFD.L", "IGLN.L"]

start = dt.date(2018, 10, 18)
end = dt.date(2022, 10, 18)

df = yf.download(tickers, start, end, interval="1d")['Adj Close']
df[["ISFD.L_ret", "IGLN.L_ret"]] = np.log(df) - np.log(df.shift(1))
df = df.iloc[1:, 2:]
```

```python
# Cheking the changes in price over the last 52weeks
df["2021-10-18":"2022-10-18"].plot();
```

The Above graph shows the variation in the daily returns of the chosen assets over the last 52 weeks. However, for better comparison, it is better to have booth asset within similar limit, so that we can fully understand the impact of volatility in prices.

```python
ax1 = df.plot(figsize=(15, 3), y="IGLN.L_ret", title="Figure 1: Daily Returns of IGNL.L")
ax2 = df.plot(figsize=(15, 3), y="ISFD.L_ret", title="Daily Returns of ISFD.l")

# setting limits
ax1.set_ylim(-0.4, 0.4);
ax1.set_ylim(-0.4, 0.4);
```

### COMPARING BOTH ASSETS BASED ON STANDARD DEVIATION AND DAILY PERCENTAGE CHANGE

```python
# Pct change
pct_daily_return = 100 * df.mean()
pct_daily_return = pd.DataFrame(pct_daily_return, columns=["PctDailyReturns"])

# Standard deviation
std_dev = df.std()
std_dev = pd.DataFrame(std_dev, columns=["StandardDeviation"])

print(pct_daily_return)
print()
print(std_dev)
```

These metrics show ISFD.L (FTSE 100) to be a better investment with a higher daily return and lower volatility than IGLN.L (Physical Gold)
